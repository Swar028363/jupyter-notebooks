```python
import pandas as pd
from sklearn.datasets import load_wine
```

```python
wine = load_wine(as_frame=True)
winedf = wine.frame
```

```python
winedf
```

```python
winedf['target'].value_counts()
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
n_bins = 3
en1 = KBinsDiscretizer(n_bins=n_bins, encode='onehot', strategy='uniform')
en2 = KBinsDiscretizer(n_bins=n_bins, encode='onehot', strategy='quantile')
```

```python
x = winedf[['proline', 'magnesium']]
```

```python
from sklearn.compose import ColumnTransformer
```

```python
t = [('first', en1, [0]), ('secound', en2, [1])]
```

```python
col_t = ColumnTransformer(transformers=t)
```

```python
col_t1 = col_t.fit_transform(x)
```

```python
col_t1
```

```python
df3 = pd.DataFrame(col_t1, columns=['pro_1', 'pro_2', 'pro_3', 'mag_1', 'mag_2', 'mag_3'])
```

```python
df4 = winedf.join(df3)
```
